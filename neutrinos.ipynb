```python
import pandas as pd

df = pd.read_csv("Neutrino Sim Data 2023 - Sheet1.csv")
df
```

```python
distance = df["Detector location (km)"]
neutrinos = df["Neutrinos (#)"]
muons = df["Muons (#)"]
electrons = df["Electrons (#)"]
```

```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots()
fig.set_size_inches(10,10)
ax.scatter(distance, neutrinos, label="neutrinos")
ax.scatter(distance, muons, label="muons")
ax.scatter(distance, electrons, label="electrons")
leg = ax.legend()
plt.xlabel("DISTANCE")
plt.ylabel("NO. PARTICLES")
```
